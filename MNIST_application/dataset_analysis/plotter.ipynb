```python
import matplotlib.pyplot as plt
import numpy as np
import itertools

with open('brute_force_acc.txt', 'r') as f:
    lines = f.readlines()
```

```python
# Given accuracy values
training_accuracies = list(map(float, lines[10].split(", ")))
inferencing_accuracies = list(map(float, lines[12].split(", ")))

# Plot histograms
plt.figure(figsize=(14, 6))

# Training accuracy histogram
plt.subplot(1, 2, 1)
plt.hist(training_accuracies, bins=20, color='blue', alpha=0.7)
plt.text(66,120, 'Silhouette Score: 0.05')
plt.title('Training Accuracy Distribution, MNIST49')
plt.xlabel('Accuracy')
plt.ylabel('Frequency')

# Inferencing accuracy histogram
plt.subplot(1, 2, 2)
plt.hist(inferencing_accuracies, bins=20, color='green', alpha=0.7)
plt.title('Inferencing Accuracy Distribution, MNIST49')
plt.xlabel('Accuracy')
plt.ylabel('Frequency')

# Display plot
plt.tight_layout()
plt.show()
```
